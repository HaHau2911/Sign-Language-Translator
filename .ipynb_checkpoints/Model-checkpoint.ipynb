```python
mp_hands = mp.solutions.holistic # Holistic model
mp_drawing = mp.solutions.drawing_utils # Drawing utilities
```

```python
def mediapipe_detection(image, model):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # COLOR CONVERSION BGR 2 RGB
    image.flags.writeable = False                  # Image is no longer writeable
    results = model.process(image)                 # Make prediction
    image.flags.writeable = True                   # Image is now writeable 
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR) # COLOR COVERSION RGB 2 BGR
    return image, results
```

```python
def draw_landmarks(image, results):
    mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_hands.HAND_CONNECTIONS) # Draw left hand connections
    mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_hands.HAND_CONNECTIONS) # Draw right hand connections
```

```python
def draw_styled_landmarks(image, results):
    mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_hands.HAND_CONNECTIONS, 
                             mp_drawing.DrawingSpec(color=(121,22,76), thickness=2, circle_radius=4), 
                             mp_drawing.DrawingSpec(color=(121,44,250), thickness=2, circle_radius=2)
                             ) 
    # Draw right hand connections  
    mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_hands.HAND_CONNECTIONS, 
                             mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=4), 
                             mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
                             ) 
```

# 3. Lấy ra các giá trị keypoints

```python
def extract_keypoints(results):
    lh = np.array([[res.x, res.y, res.z] for res in results.left_hand_landmarks.landmark]).flatten() if results.left_hand_landmarks else np.zeros(21*3)
    rh = np.array([[res.x, res.y, res.z] for res in results.right_hand_landmarks.landmark]).flatten() if results.right_hand_landmarks else np.zeros(21*3)
    return np.concatenate([lh, rh])
```

```python
# Khởi tạo list sequences và labels
sequences, labels = [], []

# Lặp qua từng hành động
for action in actions:
    # Lặp qua số lượng sequence cho mỗi hành động
    for seq in range(no_sequences):
        # Khởi tạo list window để lưu trữ các frame của mỗi sequence
        window = []
        # Lặp qua từng frame của mỗi sequence
        for frame_num in range(sequence_length):
            # Load frame từ file .npy và thêm vào window
            frame = np.load(os.path.join(DATA_PATH, action, str(seq), f"{frame_num}.npy"))
            window.append(frame)
        # Thêm window vào sequences và nhãn của hành động vào labels
        sequences.append(window)
        labels.append(label_map[action])
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, LayerNormalization, Dropout
from tensorflow.keras.layers import MultiHeadAttention, TimeDistributed
from tensorflow.keras.models import Model
import numpy as np

def create_transformer_model(input_shape, output_shape, num_layers=4, d_model=128, num_heads=8, dff=512, dropout_rate=0.1):
    inputs = Input(shape=input_shape)
    x = inputs

    seq_len = input_shape[0]  # chỉnh lại index của shape

    # Embedding layer
    x = Dense(d_model, activation='relu')(x)

    # Add positional encoding
    pos_encoding = positional_encoding(seq_len, d_model)
    x = x + pos_encoding[:, :seq_len, :]

    # Transformer blocks
    for i in range(num_layers):
        # Multi-head attention
        attn_output = MultiHeadAttention(num_heads=num_heads, key_dim=d_model)(x, x)
        attn_output = Dropout(dropout_rate)(attn_output)
        attn_output = LayerNormalization(epsilon=1e-6)(attn_output + x[:, :seq_len, :])

        # Feed-forward layer
        ffn_output = TimeDistributed(Dense(dff, activation='relu'))(attn_output)
        ffn_output = TimeDistributed(Dropout(dropout_rate))(ffn_output)
        ffn_output = TimeDistributed(Dense(d_model))(ffn_output)
        ffn_output = Dropout(dropout_rate)(ffn_output)
        ffn_output = LayerNormalization(epsilon=1e-6)(ffn_output + attn_output[:, :seq_len, :])

        x = ffn_output

    # Output layer
    x = TimeDistributed(Dense(output_shape[-1], activation='softmax'))(x)
    output = Dense(output_shape[0], activation='softmax')(x[:, -1, :])  # chỉnh lại đầu ra

    # Define and compile model
    model = Model(inputs=inputs, outputs=output)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

def positional_encoding(position, d_model):
    angle_rads = get_angles(np.arange(position)[:, np.newaxis], np.arange(d_model)[np.newaxis, :], d_model)
    # apply sin to even indices in the array; 2i
    angle_rads[:, 0::2] = np.sin(angle_rads[:, 0::2])
    # apply cos to odd indices in the array; 2i+1
    angle_rads[:, 1::2] = np.cos(angle_rads[:, 1::2])
    pos_encoding = angle_rads[np.newaxis, ...]
    return tf.cast(pos_encoding, dtype=tf.float32)

def get_angles(pos, i, d_model):
    angle_rates = 1 / np.power(10000, (2 * (i//2)) / np.float32(d_model))
    return pos * angle_rates

input_shape = (60, 126)
output_shape = (60,)

model_Transformer = create_transformer_model(input_shape=input_shape, output_shape=output_shape)
model_Transformer.summary()
model_Transformer.save('Model/model_Transformer.h5')
```

```python
def model_Transformer(X_train, y_train, batch_size=128, epochs=50, validation_data=None):
    model = create_transformer_model(input_shape=X_train.shape[1:], output_shape=y_train.shape[1:])
    model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_data=validation_data)
    return model
model_Transformer_Train = model_Transformer(X_train, y_train, batch_size=128, epochs=50, validation_data=(X_test, y_test))
```

```python
#Model Subclassing

import tensorflow as tf

class MyModel(tf.keras.Model):
    def __init__(self):
        super(MyModel, self).__init__()
        self.flatten = tf.keras.layers.Flatten()
        self.dense1 = tf.keras.layers.Dense(128, activation='relu')
        self.dropout = tf.keras.layers.Dropout(0.2)
        self.dense2 = tf.keras.layers.Dense(60, activation='softmax')

    def call(self, inputs):
        x = self.flatten(inputs)
        x = self.dense1(x)
        x = self.dropout(x)
        output = self.dense2(x)
        return output

model_Sub = MyModel()
model_Sub.build(input_shape=(None, 60, 126))
model_Sub.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model_Sub.summary()
```

```python
time_steps = 60
hidden_units = 126
epochs = 30
```

```python
# Add attention layer to the deep learning network
class attention(Layer):
    def __init__(self,**kwargs):
        super(attention,self).__init__(**kwargs)
 
    def build(self,input_shape):
        self.W=self.add_weight(name='attention_weight', shape=(input_shape[-1],1), 
                               initializer='random_normal', trainable=True)
        self.b=self.add_weight(name='attention_bias', shape=(input_shape[1],1), 
                               initializer='zeros', trainable=True)        
        super(attention, self).build(input_shape)
 
    def call(self,x):
        # Alignment scores. Pass them through tanh function
        e = K.tanh(K.dot(x,self.W)+self.b)
        # Remove dimension of size 1
        e = K.squeeze(e, axis=-1)   
        # Compute the weights
        alpha = K.softmax(e)
        # Reshape to tensorFlow format
        alpha = K.expand_dims(alpha, axis=-1)
        # Compute the context vector
        context = x * alpha
        context = K.sum(context, axis=1)
        return context
```

```python
def create_RNN_with_attention(hidden_units, dense_units, input_shape, activation):
    x=Input(shape=input_shape)
    RNN_layer = SimpleRNN(hidden_units, return_sequences=True, activation=activation)(x)
    attention_layer = attention()(RNN_layer)
    outputs=Dense(dense_units, trainable=True, activation=activation)(attention_layer)
    model=Model(x,outputs)
    model.compile(loss = 'categorical_crossentropy', optimizer='adam', metrics = ['accuracy'])
    return model    

model_attention = create_RNN_with_attention(hidden_units=hidden_units, dense_units=60, 
                                  input_shape=(time_steps,126), activation='tanh')
model_attention.summary()
```
